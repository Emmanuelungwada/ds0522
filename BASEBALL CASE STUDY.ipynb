```python
def pplot(df,i):
    plt. figure(figsize = (20,5))
    plt.subplot(1,2,1)
    
    sns.histplot(x=1, data=df,kde=True)
    plt.subplot(1,2,2)
    sns.regplot(x=i, y='W', data=df)
    plt.subplot(1,2,2)
    sns.boxplot(y=i, data = df)
    plt.show()    
```
